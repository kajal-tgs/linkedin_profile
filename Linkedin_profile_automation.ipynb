```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.firefox.options import Options
import time
import pandas as pd

url = "https://www.google.com"

# Set up Firefox options
options = Options()
options.headless = False  # Set to True if you don't want to see the browser

# Start the Firefox browser with the specified options
driver = webdriver.Firefox(options=options)

# Open the Google search page
driver.get(url)

# Define the search query
search_query = 'Salesforce Developer india opentowork "gmail com" site:linkedin.com 1..100'

# Find the search input field and enter the query
search_box = driver.find_element(By.NAME, "q")
search_box.send_keys(search_query)
search_box.send_keys(Keys.RETURN)

# Wait for the search results to load (you might need to adjust the wait time)
driver.implicitly_wait(10)

# Create an empty list to store profile details
profiles = []

# Define the number of times to scroll to load more results
scroll_iterations = 20

# Scroll to load more results
for _ in range(scroll_iterations):
    # Find all elements with the specified class on the current page
    elements = driver.find_elements(By.CSS_SELECTOR, '.VwiC3b.yXK7lf.lyLwlc.yDYNvb.W8l4ac.lEBKkf')

    # Iterate through each profile
    for i, element in enumerate(elements, start=1):
        # Extract text content for the current profile
        profile_data = {"Profile Details": element.text}
        
        # Append the profile data to the list
        profiles.append(profile_data)

        # Print or save the list for each profile
        print(f"Profile {len(profiles)}:\n{profiles[-1]['Profile Details']}\n{'='*50}\n")

    # Scroll down to load more results
    driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)

    # Wait for a short time to allow content to load
    time.sleep(2)

# Create a DataFrame from the list of profiles
df = pd.DataFrame(profiles)

# Print or save the final DataFrame
print(df)

# Close the browser
driver.quit()
```

```python
from selenium import webdriver

from selenium.webdriver.common.by import By

from selenium.webdriver.common.keys import Keys

from selenium.webdriver.firefox.options import Options

import time

import pandas as pd
 
url = "https://www.google.com"
 
# Set up Firefox options

options = Options()

options.headless = False  # Set to True if you don't want to see the browser
 
# Start the Firefox browser with the specified options

driver = webdriver.Firefox(options=options)
 
# Open the Google search page

driver.get(url)
 
# Define the search query

search_query = 'Salesforce Developer india opentowork "gmail com" site:linkedin.com 1..100'
 
# Find the search input field and enter the query

search_box = driver.find_element(By.NAME, "q")

search_box.send_keys(search_query)

search_box.send_keys(Keys.RETURN)
 
# Wait for the search results to load (you might need to adjust the wait time)

driver.implicitly_wait(10)
 
# Create an empty set to store unique profile details

unique_profiles = set()
 
# Define the number of times to scroll to load more results

scroll_iterations = 50
 
# Scroll to load more results

for _ in range(scroll_iterations):

    # Find all elements with the specified class on the current page

    elements = driver.find_elements(By.CSS_SELECTOR, '.VwiC3b.yXK7lf.lyLwlc.yDYNvb.W8l4ac.lEBKkf')
 
    # Iterate through each profile

    for i, element in enumerate(elements, start=1):

        # Extract text content for the current profile

        profile_data = {"Profile Details": element.text}

        # Check if the profile is unique

        if profile_data["Profile Details"] not in unique_profiles:

            # Append the profile data to the set

            unique_profiles.add(profile_data["Profile Details"])
 
            # Print or save the profile data

            print(f"Profile {len(unique_profiles)}:\n{profile_data['Profile Details']}\n{'='*50}\n")
 
    # Scroll down to load more results

    driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
 
    # Wait for a short time to allow content to load

    time.sleep(10)
 
# Close the browser

driver.quit()
```
